```python
from tqdm import tqdm
import pandas as pd
import re
```

```python
excel_data = pd.read_excel('naver_it2.xlsx', engine ='openpyxl')
```

```python
titles = excel_data[['title']]
print(titles)
print(type(titles))
```

```python
rea = re.compile("\([^\(\)]+\)")
```

```python
li = []

for title in titles['title']:
    if (rea.search(title)):
        m = rea.search(title)
        if (m.start() < m.end()) :
            li.append(title[:m.start()] + " "+ title[m.start() + 1:m.end() -1])
    else:
        li.append(title)
```

```python
text = "(고효율 이미지 파일 형식)"
m = rea.search(text)
print(m)
```

```python
if (m.start() < m.end()) :
    text = text[m.start() + 1:m.end() -1]
```

```python
li2 = []

for data in li:
    for tok in data.split(" "):
        li2.append(tok)
```

```python
from hanja import hangul
```

```python
hangul.is_hangul("1")
```

```python
zong = []

for data in li4:
    char = data[-1:]
    if hangul.is_hangul(char) :
        if (hangul.separate(char)[2] != 0):
            zong.append("T")
        else:
            zong.append("F")
    else:
        zong.append("F")
```

```python
print(zong)
```

```python
df2 = pd.DataFrame({'A': pd.Categorical(li4),'B': "0",'C': "0",'D': "0", 'E':"NNP", 'F':"*",
                    'G': pd.Categorical(zong), 'H': pd.Categorical(li4), 'I':"*", 'J':"*", 'K':"*", 'L':"*", 'M':"*"})
df2[50:90]
```
